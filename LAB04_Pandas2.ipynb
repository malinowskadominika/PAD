Następnie:



1. Zamień order_date na format daty. Upewnij się, że w rezultacie data będzie w formacie: rok-miesiąc-dzień;
2. Sprawdź ile i jakie unikalne wartości znajdują się w kolumnie T-shirt category;
3. Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;
4. Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.
5. Znajdź zamówienia złożone między 31 grudnia 2014 a 2 sierpnia 2016.


```python
import pandas as pd 
import numpy as np
orders_df = pd.read_csv('/content/orders.csv')
orders_df.head()
```

```python
#1
orders_df['order_date'] = pd.to_datetime(orders_df['order_date']).dt.strftime('%Y-%m-%d')
```

```python
#2
print("Unikalnych wartości w kolumnie Tshirt category jest: ", len(orders_df['tshirt_category'].unique()))
print("Te wartości to: \n", orders_df['tshirt_category'].unique())
```

```python
#3
orders_df['tshirt_category'] = orders_df['tshirt_category'].apply(lambda x: str.replace(x, 'Bl ', 'Black '))
orders_df['tshirt_category'] = orders_df['tshirt_category'].apply(lambda x: str.replace(x, "Wh ", "White "))
orders_df['tshirt_category'] = orders_df['tshirt_category'].apply(lambda x: str.replace(x, "Tshirt ", "T-Shirt "))
orders_df['tshirt_category'] = orders_df['tshirt_category'].apply(lambda x: str.replace(x, "T-shirt ", "T-Shirt "))
orders_df['tshirt_category'] = orders_df['tshirt_category'].apply(lambda x: str.replace(x, "Hoodie", "X Hoodie Unisex"))
orders_df['tshirt_category'] = orders_df['tshirt_category'].apply(lambda x: str.replace(x, "Tennis Shirt", "X Tennis-Shirt Unisex"))
```

```python
#4
orders_df['tshirt_colour'] = orders_df['tshirt_category'].apply(lambda x: x.split()[0])
orders_df['tshirt_type'] = orders_df['tshirt_category'].apply(lambda x: x.split()[1])
orders_df['tshirt_size'] = orders_df['tshirt_category'].apply(lambda x: x.split()[2])
```

```python
#5
result= orders_df[(orders_df['order_date'] > '2014-12-31') & (orders_df['order_date'] < '2016-08-02')]
```

```python
#Sprawdzenie 5
result[result['order_date'] <= '2014-12-31']
```

```python
#sprawdzenie 5
result[result['order_date'] >= '2016-08-02']
```
